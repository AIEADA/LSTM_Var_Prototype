```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import glob
```

```python
def extract_data(filename_list,data_name):

    filenum = 0
    for file in filename_list:
        data = xr.open_dataset(file)
        extracted = np.asarray(data[data_name]).reshape(-1,103*120).T

        if filenum == 0:
            data_matrix = extracted
            filenum+=1
        else:
            data_matrix = np.concatenate((data_matrix,extracted),axis=-1)
            
    nanmean = np.nanmean(data_matrix)
    data_matrix = np.nan_to_num(data_matrix,nan=nanmean)
            
    return data_matrix
```

```python
year_list = np.arange(1984,1996,step=1)
z500_filename_list = []
z700_filename_list = []
qv700_filename_list = []
t700_filename_list = []
t500_filename_list = []

for year in year_list:
    z500_filename_list.append('./Datasets/Z500/Z500_'+str(year)+'_coarse.nc')
    z700_filename_list.append('./Datasets/Z700/Z700_'+str(year)+'_coarse.nc')
    qv700_filename_list.append('./Datasets/QV700/QV700_'+str(year)+'_coarse.nc')
    t500_filename_list.append('./Datasets/T500/T500_'+str(year)+'_coarse.nc')
    t700_filename_list.append('./Datasets/T700/T700_'+str(year)+'_coarse.nc')
    
z500_data_train = extract_data(z500_filename_list,'z500')
z700_data_train = extract_data(z700_filename_list,'z500')
qv700_data_train = extract_data(qv700_filename_list,'qv700')
t700_data_train = extract_data(t700_filename_list,'t700')
t500_data_train = extract_data(t500_filename_list,'t500')
```

```python
print(z500_data_train.shape)
print(z700_data_train.shape)
print(qv700_data_train.shape)
print(t700_data_train.shape)
print(t500_data_train.shape)

training_data = np.concatenate((z500_data_train,z700_data_train,qv700_data_train,t500_data_train,t700_data_train),axis=0)
```

```python
year_list = np.arange(1997,2000,step=1)
z500_filename_list = []
z700_filename_list = []
qv700_filename_list = []
t700_filename_list = []
t500_filename_list = []

for year in year_list:
    z500_filename_list.append('./Datasets/Z500/Z500_'+str(year)+'_coarse.nc')
    z700_filename_list.append('./Datasets/Z700/Z700_'+str(year)+'_coarse.nc')
    qv700_filename_list.append('./Datasets/QV700/QV700_'+str(year)+'_coarse.nc')
    t500_filename_list.append('./Datasets/T500/T500_'+str(year)+'_coarse.nc')
    t700_filename_list.append('./Datasets/T700/T700_'+str(year)+'_coarse.nc')
    
z500_data_test = extract_data(z500_filename_list,'z500')
z700_data_test = extract_data(z700_filename_list,'z500')
qv700_data_test = extract_data(qv700_filename_list,'qv700')
t700_data_test = extract_data(t700_filename_list,'t700')
t500_data_test = extract_data(t500_filename_list,'t500')
```

```python
print(z500_data_test.shape)
print(z700_data_test.shape)
print(qv700_data_test.shape)
print(t700_data_test.shape)
print(t500_data_test.shape)

testing_data = np.concatenate((z500_data_test,z700_data_test,qv700_data_test,t500_data_test,t700_data_test),axis=0)
```

```python
# Split train and test
np.save('Training_snapshots.npy',training_data)
np.save('Testing_snapshots.npy',testing_data)

data_mean = np.mean(training_data,axis=-1)
np.save('Training_mean.npy',data_mean)

# Find fluctuations
training_fluc = training_data - data_mean[:,None]
testing_fluc = testing_data - data_mean[:,None]
```

```python
print(training_fluc.shape)
print(testing_fluc.shape)
```

```python
plt.figure()
plt.imshow(data_mean.reshape(5,103,120)[4,:,:])
plt.colorbar()
plt.show()
```

### POD computation

```python
u, s, _ = np.linalg.svd(training_fluc,full_matrices=False)
```

```python
np.save('POD_Modes.npy',u)
np.save('Singular_Values.npy',s)
```

```python
plt.figure()
plt.plot(np.cumsum(s)/np.sum(s))
plt.show()
```

```python
num_modes_trunc = 20
utrunc = u[:,:num_modes_trunc]

training_coeffs = np.matmul(utrunc.T,training_fluc)
testing_coeffs = np.matmul(utrunc.T,testing_fluc)

# Check reconstruction accuracy
snapshot_num = 100
true = training_data[:,snapshot_num].reshape(103,120)

plt.figure()
plt.imshow(true)
plt.colorbar()
plt.title('true')
plt.show()


# Check reconstruction accuracy
pred = data_mean.reshape(103,120) + np.matmul(utrunc,training_coeffs)[:,snapshot_num].reshape(103,120)

plt.figure()
plt.imshow(pred)
plt.colorbar()
plt.title('predicted')
plt.show()
```

```python
plt.figure()
plt.plot(training_coeffs[0,:],label='Mode 0')
plt.plot(training_coeffs[1,:],label='Mode 1')
plt.plot(training_coeffs[2,:],label='Mode 2')
plt.plot(training_coeffs[3,:],label='Mode 3')
plt.plot(training_coeffs[4,:],label='Mode 4')
plt.legend()
plt.show()
```

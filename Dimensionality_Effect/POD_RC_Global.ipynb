```python
model_name = 'Mode_Global_RC'
mode_num = 0
mode_range = 20
```

```python
import numpy as np
np.random.seed(10)
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt
```

```python
raw_train = np.load('../Grid_LSTM/POD/train_z500_coeffs.npy').T[:,:20]
raw_test = np.load('../Grid_LSTM/POD/test_z500_coeffs.npy').T[:,:20]
```

```python
scaler = MinMaxScaler()
train_coeffs = scaler.fit_transform(raw_train)
test_coeffs = scaler.transform(raw_test)
```

```python
from reservoirpy.nodes import Reservoir, Ridge

reservoir = Reservoir(100, lr=0.5, sr=0.9)
ridge = Ridge(ridge=1e-7)

esn_model = reservoir >> ridge

esn_model = esn_model.fit(train_coeffs[:-1], train_coeffs[1:], warmup=14)
```

```python
input_window = 14
output_window = 7

train_inputs = []
train_outputs = []

i = 0
while i < train_coeffs.shape[0] - input_window - output_window:
    train_inputs.append(train_coeffs[i:i+input_window])
    train_outputs.append(train_coeffs[i+input_window:i+input_window+output_window])
    i = i + 1
    
train_inputs = np.asarray(train_inputs)
train_outputs = np.asarray(train_outputs)[:,:,mode_num:mode_num+mode_range]
```

```python
pred_outputs = []
for j in range(len(train_inputs)):
    warmup_y = esn_model.run(train_inputs[j], reset=True)
    Y_pred = np.empty((output_window, 20))
    x = warmup_y[-1].reshape(1, -1)

    for i in range(output_window):
        x = esn_model(x)
        Y_pred[i] = x
        
    pred_outputs.append(Y_pred)
```

```python
pred_outputs = np.asarray(pred_outputs)
```

```python
for mode_num in range(20):
    plt.figure()
    plt.plot(train_outputs[:,:,mode_num].flatten(),train_outputs[:,:,mode_num].flatten(),label='True')
    plt.scatter(train_outputs[:,:,mode_num].flatten(),pred_outputs[:,:,mode_num].flatten(),label='Predicted')
    plt.legend()
    plt.show()
```
